# Removing punctuation marks and stop words 

```python
stopwords = nltk.corpus.stopwords.words('english')

prob = "".join([i for i in prob if i not in string.punctuation])
print(prob)
words = nltk.tokenize.word_tokenize(prob)
print(words)
words_new = " ".join([i for i in words if i not in stopwords])
print(words_new)
```

# Python program to convert a list to string

```python

 
def listToString(s):
 
    # initialize an empty string
    str1 = ""
 
    # traverse in the string
    for ele in s:
        str1 += ele
 
    # return string
    return str1
# Driver code
#text = listToString(rating)
#print(text)

probc = listToString(words_new)
print(words_new)
```

# Importing libraries for Image masking and numpy for image to array 

```python
from os import path
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import numpy as np
```

# Reading an image 

```python
img=Image.open("C:/Users/patha/Desktop/SCIT_DSDA_SEM1/FinanceProject/bul.png")
img
```

# converting image as array for masking 

```python
prob_mask = np.array(Image.open("C:/Users/patha/Desktop/SCIT_DSDA_SEM1/FinanceProject/bul.png"))
```

```python
stopwords = set(STOPWORDS)
```

# Defining wordcloud

```python
wc = WordCloud(background_color="white", max_words=2000, mask=prob_mask,
               stopwords=stopwords, contour_width=10, contour_color='darkgreen',collocations=False)
```

# Genarating Wordcloud

```python
wc.generate(probc)
```

# Displying ImageMasked word cloud

```python
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.imshow(prob_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```
